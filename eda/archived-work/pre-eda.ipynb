# EDA work

```python
cleaned_data = pd.read_csv("../data/clean-data/cleaned_dutchess_data.csv")
cleaned_data_original = cleaned_data.copy()
```

```python
cleaned_data_original
```

```python
cleaned_data['property_value'] = pd.to_numeric(cleaned_data['property_value'], errors='coerce')
cleaned_data['debt_to_income_ratio'] = pd.to_numeric(cleaned_data['debt_to_income_ratio'], errors='coerce')
cleaned_data['interest_rate'] = pd.to_numeric(cleaned_data['interest_rate'], errors='coerce')


cleaned_data = cleaned_data.dropna(subset=['property_value'])
cleaned_data = cleaned_data[cleaned_data['property_value'] < 1500000]
cleaned_data = cleaned_data[cleaned_data['loan_amount'] < 4000000]
cleaned_data = cleaned_data[cleaned_data['income'] < 2000]
```

```python
cleaned_data.info()
```

```python
cleaned_data["loan_rejected"] = (cleaned_data['action_taken'] == 3)
cleaned_data["loan_approved"] = (cleaned_data['action_taken'] == 2)
# cleaned_data["loan_rejected"] = cleaned_data["loan_rejected"].astype(int)
```

Figure 3: This figure looks at the loans rejected and looks at the income in comparison to the property values of the houses properties being applied for to get a sense of how income effects the loan people apply for and how that effects loan status.

```python
plt.style.use('seaborn-v0_8-whitegrid')

plt.figure(figsize=(12, 10))

markers = {
    1: 'o',   
    2: 's',  
    3: 'x',   
    4: '^',   
    5: 'v',   
    6: '+',  
    7: '*',  
    8: 'D'   
}

colors = {
    1: 'green',    
    2: 'blue',   
    3: 'red',    
    4: 'orange',  
    5: 'purple', 
    6: 'brown',   
    7: 'gray',    
    8: 'cyan'     
}

# Status labels for legend
status_labels = {
    1: 'Loan Originated',
    2: 'Application Approved but Not Accepted',
    3: 'Application Denied',
    4: 'Application Withdrawn by Applicant',
    5: 'File Closed for Incompleteness',
    6: 'Purchased Loan',
    7: 'Preapproval Request Denied',
    8: 'Preapproval Request Approved but Not Accepted'
}

# Create the scatter plot with custom markers for each status
for status in range(1, 9):
    mask = cleaned_data['action_taken'] == status
    plt.scatter(
        x=cleaned_data.loc[mask, 'debt_to_income_ratio'], 
        y=cleaned_data.loc[mask, 'property_value'], 
        marker=markers[status],
        c=colors[status],
        alpha=0.7,
        edgecolors='black',
        linewidth=0.5,
        label=status_labels[status]
    )

plt.xlabel('Debt-to-income ratio', fontsize=12)
plt.ylabel('Property Value', fontsize=12)
plt.title('Debt-to-income ratio vs Property Value - Loan Action Status', fontsize=14)

# Set axes to start at 0
# plt.xlim(left=0)
plt.ylim(bottom=0)

plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.style.use('seaborn-v0_8-whitegrid')

plt.figure(figsize=(10, 10))

# Use different markers for True and False
markers = { True: 'x', False: 'o'}
colors = { True: 'red', False: 'blue'}

# Create the scatter plot with custom markers
for status in [False, True]:
    mask = cleaned_data['loan_rejected'] == status
    plt.scatter(
        x=cleaned_data.loc[mask, 'income'], 
        y=cleaned_data.loc[mask, 'property_value'], 
        marker=markers[status],
        c=colors[status],
        alpha=0.7,
        edgecolors='black',
        linewidth=0.5,
        label=f'Loan {"rejected" if status else "not rejected"}'
    )

plt.xlabel('Income', fontsize=12)
plt.ylabel('Property Value', fontsize=12)
plt.title('Fig: 3: Income vs Property Value - Loan Rejected Status', fontsize=14)

# Set y-axis to start at 0 and use very few ticks
plt.xlim(left=0)
plt.ylim(bottom=0)

# Add legend instead of colorbar
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['tract_minority_population_percent'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Tract Minority Population Percent')
plt.xlabel('Minority Population Percent')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['tract_minority_population_percent'], df['income'], alpha=0.5, color='teal', edgecolors='k')
plt.title('Scatterplot of Income vs. Tract Minority Population Percent')
plt.xlabel('Tract Minority Population Percent')
plt.ylabel('Applicant Income')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Filter out outliers using the 1st and 99th percentiles
filtered_df = df[
    (df['income'].between(df['income'].quantile(0.01), df['income'].quantile(0.99))) &
    (df['tract_minority_population_percent'].between(
        df['tract_minority_population_percent'].quantile(0.01),
        df['tract_minority_population_percent'].quantile(0.99)
    ))
]

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['tract_minority_population_percent'], filtered_df['income'],
            alpha=0.5, color='teal', edgecolors='k')
plt.title('Scatterplot of Income vs. Tract Minority Population Percent (Outliers Removed)')
plt.xlabel('Tract Minority Population Percent')
plt.ylabel('Applicant Income')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hexbin(filtered_df['tract_minority_population_percent'], filtered_df['income'], gridsize=40, cmap='viridis', mincnt=1)
plt.colorbar(label='Count in bin')
plt.xlabel('Tract Minority Population Percent')
plt.ylabel('Applicant Income')
plt.title('Hexbin Plot: Income vs. Minority Population Percent')
plt.tight_layout()
plt.show()
```
